```python
import time
from pathlib import Path
import duckdb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from scipy.stats import zscore
import matplotlib as mpl
from matplotlib.colors import LinearSegmentedColormap

start_time = time.time()

# 设置全局可视化风格 - 高级好看的配置
plt.style.use('seaborn-v0_8-white')  # 基础白色风格

# 设置背景和边框
mpl.rcParams['figure.facecolor'] = 'white'
mpl.rcParams['figure.edgecolor'] = 'black'
mpl.rcParams['figure.figsize'] = (10, 6)
mpl.rcParams['axes.facecolor'] = '#f8f9fa'
mpl.rcParams['axes.edgecolor'] = 'black'
mpl.rcParams['axes.linewidth'] = 1.5
mpl.rcParams['axes.grid'] = True
mpl.rcParams['grid.color'] = '#e0e0e0'
mpl.rcParams['grid.linestyle'] = '--'
mpl.rcParams['grid.linewidth'] = 0.8
mpl.rcParams['grid.alpha'] = 0.5

# 设置字体
mpl.rcParams['font.sans-serif'] = ['Microsoft YaHei']
mpl.rcParams['font.size'] = 11
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['axes.titlesize'] = 16
mpl.rcParams['axes.titleweight'] = 'bold'
mpl.rcParams['axes.labelsize'] = 12
mpl.rcParams['xtick.labelsize'] = 10
mpl.rcParams['ytick.labelsize'] = 10

# 设置线条和点的样式
mpl.rcParams['lines.linewidth'] = 2.5
mpl.rcParams['patch.edgecolor'] = 'black'
mpl.rcParams['patch.linewidth'] = 1

# 自定义高级配色方案
pastel_colors = ['#c6dbef', '#9ecae1', '#6baed6', '#4292c6', '#2171b5', '#084594']
warm_colors = ['#fee6ce', '#fdd0a2', '#fdae6b', '#fd8d3c', '#f16913', '#d94801']
elegant_palette = sns.color_palette(pastel_colors)
warm_palette = sns.color_palette(warm_colors)

# 数据加载参数
data_dir = Path(r"E:\BaiduNetdiskDownload\10G_data_new")
glob_pat = str(data_dir / "part-0000*.parquet")
ref_date = "2025-02-01"

con = duckdb.connect()

# 预览前 5 行
preview_df = con.sql(f"SELECT * FROM '{glob_pat}' LIMIT 5").df()
print("=== Preview (first 5 rows) ===")
print(preview_df)

# 缺失值统计
total_rows = con.sql(f"SELECT COUNT(*) FROM '{glob_pat}'").fetchone()[0]
cols = [row[0] for row in con.sql(f"DESCRIBE SELECT * FROM '{glob_pat}'").fetchall()]

stats = []
for col in cols:
    non_null = con.sql(f'SELECT COUNT("{col}") FROM "{glob_pat}"').fetchone()[0]
    missing = total_rows - non_null
    stats.append((col, missing, missing / total_rows))

missing_summary = (pd.DataFrame(stats, columns=["column", "missing_count", "missing_ratio"])
                  .sort_values("missing_ratio", ascending=False))

print("\n=== Missing-value summary (top 10 by ratio) ===")
print(missing_summary.head(10))

# 数据提取与转换
query = f"""
SELECT
    id,
    age,
    income,
    CAST(json_extract(purchase_history, '$.avg_price') AS DOUBLE) AS avg_price,
    json_array_length(json_extract(purchase_history, '$.items')) AS item_count,
    DATE_DIFF('day',
             CAST(json_extract(purchase_history, '$.purchase_date') AS DATE),
             DATE '{ref_date}') AS days_since_last_purchase,
    CAST(json_extract(purchase_history, '$.avg_price') AS DOUBLE) / income AS income_spend_ratio
FROM '{glob_pat}'
"""

# 获取查询结果
result_df = con.sql(query).df()

# 显示结果前几行
print("\n=== Transformed data (first 5 rows) ===")
print(result_df.head())

# 描述性统计
description = result_df.describe()
print("\n=== Descriptive Statistics ===")
print(description)

# 异常值检测
numeric_columns = ['age', 'income', 'avg_price', 'item_count', 'days_since_last_purchase', 'income_spend_ratio']
z_scores = result_df[numeric_columns].apply(zscore)
outliers = (np.abs(z_scores) > 3)

print("\n=== Outlier Count (based on Z-score) ===")
print(outliers.sum())

# 去除异常值
result_df_no_outliers = result_df[~(outliers.any(axis=1))]
print(f"\n=== Data after removing outliers: {len(result_df_no_outliers)} rows (removed {len(result_df) - len(result_df_no_outliers)} outliers) ===")

# 相关性分析 - 高级美观热力图
numeric_columns = result_df.drop(columns=['id']).select_dtypes(include=['float64', 'int64'])
corr_matrix = numeric_columns.corr()

# 创建自定义渐变色映射
cmap = LinearSegmentedColormap.from_list('elegant_cmap', ['#f7fbff', '#3182bd'])

fig, ax = plt.subplots(figsize=(11, 9))
fig.patch.set_facecolor('white')
fig.patch.set_edgecolor('black')
fig.patch.set_linewidth(2)

mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
heatmap = sns.heatmap(
    corr_matrix,
    mask=mask,
    annot=True,
    cmap=cmap,
    fmt='.6f',
    linewidths=1,
    cbar_kws={'shrink': 0.8, 'label': '相关系数'},
    annot_kws={'size': 10, 'weight': 'bold'}
)

# 添加边框和网格线以提升专业感
for _, spine in ax.spines.items():
    spine.set_visible(True)
    spine.set_color('black')
    spine.set_linewidth(1.5)

plt.title('特征相关性分析', fontsize=18, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

# 年龄段分析 - 时尚条形图
age_bins = [18, 29, 44, 59, 100]
age_labels = ['18-29', '30-44', '45-59', '60+']
result_df['age_group'] = pd.cut(result_df['age'], bins=age_bins, labels=age_labels, right=True)
avg_price_by_age_group = result_df.groupby('age_group')['avg_price'].mean().reset_index()

fig, ax = plt.subplots(figsize=(10, 7))
fig.patch.set_facecolor('white')
fig.patch.set_edgecolor('black')
fig.patch.set_linewidth(2)

# 创建渐变色调色板
gradient_colors = sns.color_palette("Blues", n_colors=len(age_labels) + 2)[1:-1]

bars = sns.barplot(
    x='age_group',
    y='avg_price',
    data=avg_price_by_age_group,
    palette=gradient_colors,
    edgecolor='black',
    linewidth=1.5,
    ax=ax
)

# 添加数据标签，使用阴影效果增强专业感
for i, bar in enumerate(bars.patches):
    height = bar.get_height()
    bars.text(
        bar.get_x() + bar.get_width()/2.,
        height + 5,
        f'{height:.2f}',
        ha='center',
        va='bottom',
        fontsize=12,
        fontweight='bold',
        color='#2c3e50',
        bbox=dict(boxstyle="round,pad=0.3", facecolor='white', edgecolor='#d4d4d4', alpha=0.9)
    )

# 添加精致的网格线
ax.yaxis.grid(True, linestyle='--', alpha=0.7, color='#d4d4d4')
ax.set_axisbelow(True)

# 美化边框
for spine in ax.spines.values():
    spine.set_color('black')
    spine.set_linewidth(1.5)

plt.title('各年龄段平均商品价格分析', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('年龄段', fontsize=14, labelpad=10)
plt.ylabel('平均商品价格', fontsize=14, labelpad=10)
plt.tight_layout()
plt.show()

# 分布图 - 高端双直方图
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.patch.set_facecolor('white')
fig.patch.set_edgecolor('black')
fig.patch.set_linewidth(2)

# 年龄分布 - 使用渐变色填充
n, bins, patches = axes[0].hist(
    result_df['age'],
    bins=20,
    edgecolor='black',
    linewidth=1.2,
    alpha=0.8
)

# 为直方图添加渐变色
cm = plt.cm.Blues
bin_centers = 0.5 * (bins[:-1] + bins[1:])
col = bin_centers - min(bin_centers)
col /= max(col)
for c, p in zip(col, patches):
    plt.setp(p, 'facecolor', cm(c * 0.8 + 0.2))

# 添加平滑曲线以增强专业感
sns.kdeplot(result_df['age'], ax=axes[0], color='#d62728', linewidth=2.5)

axes[0].set_title('客户年龄分布', fontsize=16, fontweight='bold', pad=15)
axes[0].set_xlabel('年龄', fontsize=13, labelpad=10)
axes[0].set_ylabel('频率', fontsize=13, labelpad=10)
axes[0].grid(True, linestyle='--', alpha=0.7, color='#d4d4d4')
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)
axes[0].spines['left'].set_linewidth(1.5)
axes[0].spines['bottom'].set_linewidth(1.5)

# 平均价格分布 - 使用渐变色填充
n, bins, patches = axes[1].hist(
    result_df['avg_price'],
    bins=20,
    edgecolor='black',
    linewidth=1.2,
    alpha=0.8
)

# 为直方图添加渐变色
cm = plt.cm.Reds
bin_centers = 0.5 * (bins[:-1] + bins[1:])
col = bin_centers - min(bin_centers)
col /= max(col)
for c, p in zip(col, patches):
    plt.setp(p, 'facecolor', cm(c * 0.8 + 0.2))

# 添加平滑曲线以增强专业感
sns.kdeplot(result_df['avg_price'], ax=axes[1], color='#1f77b4', linewidth=2.5)

axes[1].set_title('平均购买价格分布', fontsize=16, fontweight='bold', pad=15)
axes[1].set_xlabel('平均价格', fontsize=13, labelpad=10)
axes[1].set_ylabel('频率', fontsize=13, labelpad=10)
axes[1].grid(True, linestyle='--', alpha=0.7, color='#d4d4d4')
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)
axes[1].spines['left'].set_linewidth(1.5)
axes[1].spines['bottom'].set_linewidth(1.5)

plt.tight_layout()
plt.subplots_adjust(wspace=0.25)
plt.show()

# 聚类分析
features = ['age', 'avg_price', 'item_count', 'days_since_last_purchase', 'income', 'income_spend_ratio']
data_for_clustering = result_df[features]

scaler = MinMaxScaler()
data_normalized = scaler.fit_transform(data_for_clustering)

kmeans = KMeans(n_clusters=3, random_state=42)
result_df['cluster'] = kmeans.fit_predict(data_normalized)

# 聚类饼图 - 高级设计
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(aspect="equal"))
fig.patch.set_facecolor('white')
fig.patch.set_edgecolor('black')
fig.patch.set_linewidth(2)

cluster_counts = result_df['cluster'].value_counts()

# 优雅的配色方案
elegant_colors = ['#8dd3c7', '#bebada', '#fb8072']

# 绘制饼图
wedges, texts, autotexts = ax.pie(
    cluster_counts,
    labels=None,  # 先不添加标签，后面自定义
    autopct='%1.1f%%',
    startangle=90,
    colors=elegant_colors,
    wedgeprops={'edgecolor': 'black', 'linewidth': 1.5, 'antialiased': True},
    textprops={'fontsize': 12, 'fontweight': 'bold'},
    shadow=True,
    explode=(0.05, 0.05, 0.05)  # 轻微分离所有部分
)

# 自定义百分比文本样式
for autotext in autotexts:
    autotext.set_fontsize(13)
    autotext.set_fontweight('bold')
    autotext.set_color('black')

# 添加精美的图例
legend_labels = [f'类别 {i}\n({count}个样本)' for i, count in cluster_counts.items()]
ax.legend(
    wedges,
    legend_labels,
    title="聚类分布",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1),
    frameon=True,
    facecolor='white',
    edgecolor='black',
    title_fontsize=14,
    fontsize=12
)

plt.title('用户聚类分布分析', fontsize=18, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

# 聚类特征对比 - 高级条形图
cluster_means = result_df.groupby('cluster')[features].mean()

fig, axes = plt.subplots(2, 3, figsize=(16, 10))
fig.patch.set_facecolor('white')
fig.patch.set_edgecolor('black')
fig.patch.set_linewidth(2)

# 创建优雅的颜色方案
color_palette = ['#8dd3c7', '#bebada', '#fb8072']

# 自定义特征名称映射，使其更专业
feature_names = {
    'age': '年龄',
    'avg_price': '平均价格',
    'item_count': '商品数量',
    'days_since_last_purchase': '最近购买天数',
    'income': '收入',
    'income_spend_ratio': '收入支出比'
}

for i, feature in enumerate(features):
    ax = axes[i // 3, i % 3]
    ax.set_facecolor('#f8f9fa')

    # 绘制条形图
    bars = sns.barplot(
        x=cluster_means.index,
        y=feature,
        data=cluster_means,
        palette=color_palette,
        ax=ax,
        edgecolor='black',
        linewidth=1.2
    )

    # 在条形图上添加数值标注，加上专业设计
    for j, p in enumerate(ax.patches):
        height = p.get_height()
        ax.annotate(
            f'{height:.2f}',
            (p.get_x() + p.get_width() / 2., height),
            ha='center',
            va='bottom',
            fontsize=10,
            fontweight='bold',
            bbox=dict(boxstyle="round,pad=0.3", facecolor='white', alpha=0.8, edgecolor='#d4d4d4')
        )

    # 设置图表样式
    ax.set_title(f'{feature_names.get(feature, feature)}比较', fontsize=14, fontweight='bold', pad=15)
    ax.set_xlabel('聚类', fontsize=12, labelpad=10)
    ax.set_ylabel('平均值', fontsize=12, labelpad=10)
    ax.set_xticklabels([f'类别 {i}' for i in range(3)], fontsize=10)

    # 添加网格线
    ax.yaxis.grid(True, linestyle='--', alpha=0.6, color='#d4d4d4')
    ax.set_axisbelow(True)

    # 美化边框
    for spine in ax.spines.values():
        spine.set_color('black')
        spine.set_linewidth(1)

# 调整子图之间的间距
plt.tight_layout()
plt.subplots_adjust(hspace=0.3, wspace=0.25)
plt.show()

# 雷达图 - 高级设计
labels = [feature_names.get(feature, feature) for feature in features]
num_vars = len(labels)

cluster_averages_normalized = scaler.transform(cluster_means)

# 修复：确保角度和值的长度相同
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # 添加第一个角度到末尾以闭合图形

fig = plt.figure(figsize=(12, 10))
fig.patch.set_facecolor('white')
fig.patch.set_edgecolor('black')
fig.patch.set_linewidth(2)

# 设置背景为浅灰色并添加圆形边框
ax = fig.add_subplot(111, polar=True, facecolor='#f8f9fa')

# 自定义雷达图颜色和样式
radar_colors = ['#3182bd', '#9ecae1', '#fdae6b']
line_styles = ['-', '--', '-.']
alphas = [0.3, 0.25, 0.2]

# 绘制多边形网格线
for i in range(num_vars):
    angle = angles[i]
    ax.plot([angle, angle], [0, 1.3], color="#d4d4d4", linestyle='-', linewidth=1, alpha=0.8)

# 绘制同心圆
for r in np.linspace(0, 1, 5):
    circle = plt.Circle((0, 0), r, transform=ax.transData._b, fill=False, color="#d4d4d4", linestyle='-', linewidth=1, alpha=0.6)
    ax.add_artist(circle)

# 绘制每个聚类的雷达图
for i in range(3):
    values = cluster_averages_normalized[i].tolist()
    values += values[:1]  # 将第一个值添加到最后一个，以闭合图形

    # 绘制线条
    ax.plot(
        angles,
        values,
        linewidth=3,
        linestyle=line_styles[i],
        color=radar_colors[i],
        label=f'类别 {i}',
        zorder=10
    )

    # 填充区域
    ax.fill(
        angles,
        values,
        color=radar_colors[i],
        alpha=alphas[i],
        zorder=5
    )

# 设置雷达图标签样式
ax.set_xticks(angles[:-1])
ax.set_xticklabels(
    labels,
    fontsize=12,
    fontweight='bold',
    color='#2c3e50'
)

# 移除刻度线和标签
ax.set_yticklabels([])
ax.set_ylim(0, 1.2)

# 添加精美的图例
legend = plt.legend(
    loc='upper right',
    bbox_to_anchor=(1.2, 1.1),
    frameon=True,
    facecolor='white',
    edgecolor='black',
    fontsize=12
)
legend.get_frame().set_linewidth(1.5)

plt.title('聚类特征雷达分析', fontsize=18, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

end_time = time.time()
execution_time = end_time - start_time
print(f"\n=== 脚本执行时间: {execution_time:.2f} 秒 ===")
```

```python
import time
from pathlib import Path
import duckdb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from scipy.stats import zscore
import matplotlib as mpl
from matplotlib.colors import LinearSegmentedColormap

start_time = time.time()

# 设置全局可视化风格 - 高级好看的配置
plt.style.use('seaborn-v0_8-white')  # 基础白色风格

# 设置背景和边框
mpl.rcParams['figure.facecolor'] = 'white'
mpl.rcParams['figure.edgecolor'] = 'black'
mpl.rcParams['figure.figsize'] = (10, 6)
mpl.rcParams['axes.facecolor'] = '#f8f9fa'
mpl.rcParams['axes.edgecolor'] = 'black'
mpl.rcParams['axes.linewidth'] = 1.5
mpl.rcParams['axes.grid'] = True
mpl.rcParams['grid.color'] = '#e0e0e0'
mpl.rcParams['grid.linestyle'] = '--'
mpl.rcParams['grid.linewidth'] = 0.8
mpl.rcParams['grid.alpha'] = 0.5

# 设置字体
mpl.rcParams['font.sans-serif'] = ['Microsoft YaHei']
mpl.rcParams['font.size'] = 11
mpl.rcParams['axes.unicode_minus'] = False
mpl.rcParams['axes.titlesize'] = 16
mpl.rcParams['axes.titleweight'] = 'bold'
mpl.rcParams['axes.labelsize'] = 12
mpl.rcParams['xtick.labelsize'] = 10
mpl.rcParams['ytick.labelsize'] = 10

# 设置线条和点的样式
mpl.rcParams['lines.linewidth'] = 2.5
mpl.rcParams['patch.edgecolor'] = 'black'
mpl.rcParams['patch.linewidth'] = 1

# 自定义高级配色方案
pastel_colors = ['#c6dbef', '#9ecae1', '#6baed6', '#4292c6', '#2171b5', '#084594']
warm_colors = ['#fee6ce', '#fdd0a2', '#fdae6b', '#fd8d3c', '#f16913', '#d94801']
elegant_palette = sns.color_palette(pastel_colors)
warm_palette = sns.color_palette(warm_colors)

# 数据加载参数
data_dir = Path(r"E:\BaiduNetdiskDownload\30G_data_new")
glob_pat = str(data_dir / "part-0000*.parquet")
ref_date = "2025-02-01"

con = duckdb.connect()

# 预览前 5 行
preview_df = con.sql(f"SELECT * FROM '{glob_pat}' LIMIT 5").df()
print("=== Preview (first 5 rows) ===")
print(preview_df)

# 缺失值统计
total_rows = con.sql(f"SELECT COUNT(*) FROM '{glob_pat}'").fetchone()[0]
cols = [row[0] for row in con.sql(f"DESCRIBE SELECT * FROM '{glob_pat}'").fetchall()]

stats = []
for col in cols:
    non_null = con.sql(f'SELECT COUNT("{col}") FROM "{glob_pat}"').fetchone()[0]
    missing = total_rows - non_null
    stats.append((col, missing, missing / total_rows))

missing_summary = (pd.DataFrame(stats, columns=["column", "missing_count", "missing_ratio"])
                  .sort_values("missing_ratio", ascending=False))

print("\n=== Missing-value summary (top 10 by ratio) ===")
print(missing_summary.head(10))

# 数据提取与转换
query = f"""
SELECT
    id,
    age,
    income,
    CAST(json_extract(purchase_history, '$.avg_price') AS DOUBLE) AS avg_price,
    json_array_length(json_extract(purchase_history, '$.items')) AS item_count,
    DATE_DIFF('day',
             CAST(json_extract(purchase_history, '$.purchase_date') AS DATE),
             DATE '{ref_date}') AS days_since_last_purchase,
    CAST(json_extract(purchase_history, '$.avg_price') AS DOUBLE) / income AS income_spend_ratio
FROM '{glob_pat}'
"""

# 获取查询结果
result_df = con.sql(query).df()

# 显示结果前几行
print("\n=== Transformed data (first 5 rows) ===")
print(result_df.head())

# 描述性统计
description = result_df.describe()
print("\n=== Descriptive Statistics ===")
print(description)

# 异常值检测
numeric_columns = ['age', 'income', 'avg_price', 'item_count', 'days_since_last_purchase', 'income_spend_ratio']
z_scores = result_df[numeric_columns].apply(zscore)
outliers = (np.abs(z_scores) > 3)

print("\n=== Outlier Count (based on Z-score) ===")
print(outliers.sum())

# 去除异常值
result_df_no_outliers = result_df[~(outliers.any(axis=1))]
print(f"\n=== Data after removing outliers: {len(result_df_no_outliers)} rows (removed {len(result_df) - len(result_df_no_outliers)} outliers) ===")

# 相关性分析 - 高级美观热力图
numeric_columns = result_df.drop(columns=['id']).select_dtypes(include=['float64', 'int64'])
corr_matrix = numeric_columns.corr()

# 创建自定义渐变色映射
cmap = LinearSegmentedColormap.from_list('elegant_cmap', ['#f7fbff', '#3182bd'])

fig, ax = plt.subplots(figsize=(11, 9))
fig.patch.set_facecolor('white')
fig.patch.set_edgecolor('black')
fig.patch.set_linewidth(2)

mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
heatmap = sns.heatmap(
    corr_matrix,
    mask=mask,
    annot=True,
    cmap=cmap,
    fmt='.6f',
    linewidths=1,
    cbar_kws={'shrink': 0.8, 'label': '相关系数'},
    annot_kws={'size': 10, 'weight': 'bold'}
)

# 添加边框和网格线以提升专业感
for _, spine in ax.spines.items():
    spine.set_visible(True)
    spine.set_color('black')
    spine.set_linewidth(1.5)

plt.title('特征相关性分析', fontsize=18, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

# 年龄段分析 - 时尚条形图
age_bins = [18, 29, 44, 59, 100]
age_labels = ['18-29', '30-44', '45-59', '60+']
result_df['age_group'] = pd.cut(result_df['age'], bins=age_bins, labels=age_labels, right=True)
avg_price_by_age_group = result_df.groupby('age_group')['avg_price'].mean().reset_index()

fig, ax = plt.subplots(figsize=(10, 7))
fig.patch.set_facecolor('white')
fig.patch.set_edgecolor('black')
fig.patch.set_linewidth(2)

# 创建渐变色调色板
gradient_colors = sns.color_palette("Blues", n_colors=len(age_labels) + 2)[1:-1]

bars = sns.barplot(
    x='age_group',
    y='avg_price',
    data=avg_price_by_age_group,
    palette=gradient_colors,
    edgecolor='black',
    linewidth=1.5,
    ax=ax
)

# 添加数据标签，使用阴影效果增强专业感
for i, bar in enumerate(bars.patches):
    height = bar.get_height()
    bars.text(
        bar.get_x() + bar.get_width()/2.,
        height + 5,
        f'{height:.2f}',
        ha='center',
        va='bottom',
        fontsize=12,
        fontweight='bold',
        color='#2c3e50',
        bbox=dict(boxstyle="round,pad=0.3", facecolor='white', edgecolor='#d4d4d4', alpha=0.9)
    )

# 添加精致的网格线
ax.yaxis.grid(True, linestyle='--', alpha=0.7, color='#d4d4d4')
ax.set_axisbelow(True)

# 美化边框
for spine in ax.spines.values():
    spine.set_color('black')
    spine.set_linewidth(1.5)

plt.title('各年龄段平均商品价格分析', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('年龄段', fontsize=14, labelpad=10)
plt.ylabel('平均商品价格', fontsize=14, labelpad=10)
plt.tight_layout()
plt.show()

# 分布图 - 高端双直方图
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.patch.set_facecolor('white')
fig.patch.set_edgecolor('black')
fig.patch.set_linewidth(2)

# 年龄分布 - 使用渐变色填充
n, bins, patches = axes[0].hist(
    result_df['age'],
    bins=20,
    edgecolor='black',
    linewidth=1.2,
    alpha=0.8
)

# 为直方图添加渐变色
cm = plt.cm.Blues
bin_centers = 0.5 * (bins[:-1] + bins[1:])
col = bin_centers - min(bin_centers)
col /= max(col)
for c, p in zip(col, patches):
    plt.setp(p, 'facecolor', cm(c * 0.8 + 0.2))

# 添加平滑曲线以增强专业感
sns.kdeplot(result_df['age'], ax=axes[0], color='#d62728', linewidth=2.5)

axes[0].set_title('客户年龄分布', fontsize=16, fontweight='bold', pad=15)
axes[0].set_xlabel('年龄', fontsize=13, labelpad=10)
axes[0].set_ylabel('频率', fontsize=13, labelpad=10)
axes[0].grid(True, linestyle='--', alpha=0.7, color='#d4d4d4')
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)
axes[0].spines['left'].set_linewidth(1.5)
axes[0].spines['bottom'].set_linewidth(1.5)

# 平均价格分布 - 使用渐变色填充
n, bins, patches = axes[1].hist(
    result_df['avg_price'],
    bins=20,
    edgecolor='black',
    linewidth=1.2,
    alpha=0.8
)

# 为直方图添加渐变色
cm = plt.cm.Reds
bin_centers = 0.5 * (bins[:-1] + bins[1:])
col = bin_centers - min(bin_centers)
col /= max(col)
for c, p in zip(col, patches):
    plt.setp(p, 'facecolor', cm(c * 0.8 + 0.2))

# 添加平滑曲线以增强专业感
sns.kdeplot(result_df['avg_price'], ax=axes[1], color='#1f77b4', linewidth=2.5)

axes[1].set_title('平均购买价格分布', fontsize=16, fontweight='bold', pad=15)
axes[1].set_xlabel('平均价格', fontsize=13, labelpad=10)
axes[1].set_ylabel('频率', fontsize=13, labelpad=10)
axes[1].grid(True, linestyle='--', alpha=0.7, color='#d4d4d4')
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)
axes[1].spines['left'].set_linewidth(1.5)
axes[1].spines['bottom'].set_linewidth(1.5)

plt.tight_layout()
plt.subplots_adjust(wspace=0.25)
plt.show()

# 聚类分析
features = ['age', 'avg_price', 'item_count', 'days_since_last_purchase', 'income', 'income_spend_ratio']
data_for_clustering = result_df[features]

scaler = MinMaxScaler()
data_normalized = scaler.fit_transform(data_for_clustering)

kmeans = KMeans(n_clusters=3, random_state=42)
result_df['cluster'] = kmeans.fit_predict(data_normalized)

# 聚类饼图 - 高级设计
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(aspect="equal"))
fig.patch.set_facecolor('white')
fig.patch.set_edgecolor('black')
fig.patch.set_linewidth(2)

cluster_counts = result_df['cluster'].value_counts()

# 优雅的配色方案
elegant_colors = ['#8dd3c7', '#bebada', '#fb8072']

# 绘制饼图
wedges, texts, autotexts = ax.pie(
    cluster_counts,
    labels=None,  # 先不添加标签，后面自定义
    autopct='%1.1f%%',
    startangle=90,
    colors=elegant_colors,
    wedgeprops={'edgecolor': 'black', 'linewidth': 1.5, 'antialiased': True},
    textprops={'fontsize': 12, 'fontweight': 'bold'},
    shadow=True,
    explode=(0.05, 0.05, 0.05)  # 轻微分离所有部分
)

# 自定义百分比文本样式
for autotext in autotexts:
    autotext.set_fontsize(13)
    autotext.set_fontweight('bold')
    autotext.set_color('black')

# 添加精美的图例
legend_labels = [f'类别 {i}\n({count}个样本)' for i, count in cluster_counts.items()]
ax.legend(
    wedges,
    legend_labels,
    title="聚类分布",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1),
    frameon=True,
    facecolor='white',
    edgecolor='black',
    title_fontsize=14,
    fontsize=12
)

plt.title('用户聚类分布分析', fontsize=18, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

# 聚类特征对比 - 高级条形图
cluster_means = result_df.groupby('cluster')[features].mean()

fig, axes = plt.subplots(2, 3, figsize=(16, 10))
fig.patch.set_facecolor('white')
fig.patch.set_edgecolor('black')
fig.patch.set_linewidth(2)

# 创建优雅的颜色方案
color_palette = ['#8dd3c7', '#bebada', '#fb8072']

# 自定义特征名称映射，使其更专业
feature_names = {
    'age': '年龄',
    'avg_price': '平均价格',
    'item_count': '商品数量',
    'days_since_last_purchase': '最近购买天数',
    'income': '收入',
    'income_spend_ratio': '收入支出比'
}

for i, feature in enumerate(features):
    ax = axes[i // 3, i % 3]
    ax.set_facecolor('#f8f9fa')

    # 绘制条形图
    bars = sns.barplot(
        x=cluster_means.index,
        y=feature,
        data=cluster_means,
        palette=color_palette,
        ax=ax,
        edgecolor='black',
        linewidth=1.2
    )

    # 在条形图上添加数值标注，加上专业设计
    for j, p in enumerate(ax.patches):
        height = p.get_height()
        ax.annotate(
            f'{height:.2f}',
            (p.get_x() + p.get_width() / 2., height),
            ha='center',
            va='bottom',
            fontsize=10,
            fontweight='bold',
            bbox=dict(boxstyle="round,pad=0.3", facecolor='white', alpha=0.8, edgecolor='#d4d4d4')
        )

    # 设置图表样式
    ax.set_title(f'{feature_names.get(feature, feature)}比较', fontsize=14, fontweight='bold', pad=15)
    ax.set_xlabel('聚类', fontsize=12, labelpad=10)
    ax.set_ylabel('平均值', fontsize=12, labelpad=10)
    ax.set_xticklabels([f'类别 {i}' for i in range(3)], fontsize=10)

    # 添加网格线
    ax.yaxis.grid(True, linestyle='--', alpha=0.6, color='#d4d4d4')
    ax.set_axisbelow(True)

    # 美化边框
    for spine in ax.spines.values():
        spine.set_color('black')
        spine.set_linewidth(1)

# 调整子图之间的间距
plt.tight_layout()
plt.subplots_adjust(hspace=0.3, wspace=0.25)
plt.show()

# 雷达图 - 高级设计
labels = [feature_names.get(feature, feature) for feature in features]
num_vars = len(labels)

cluster_averages_normalized = scaler.transform(cluster_means)

# 修复：确保角度和值的长度相同
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # 添加第一个角度到末尾以闭合图形

fig = plt.figure(figsize=(12, 10))
fig.patch.set_facecolor('white')
fig.patch.set_edgecolor('black')
fig.patch.set_linewidth(2)

# 设置背景为浅灰色并添加圆形边框
ax = fig.add_subplot(111, polar=True, facecolor='#f8f9fa')

# 自定义雷达图颜色和样式
radar_colors = ['#3182bd', '#9ecae1', '#fdae6b']
line_styles = ['-', '--', '-.']
alphas = [0.3, 0.25, 0.2]

# 绘制多边形网格线
for i in range(num_vars):
    angle = angles[i]
    ax.plot([angle, angle], [0, 1.3], color="#d4d4d4", linestyle='-', linewidth=1, alpha=0.8)

# 绘制同心圆
for r in np.linspace(0, 1, 5):
    circle = plt.Circle((0, 0), r, transform=ax.transData._b, fill=False, color="#d4d4d4", linestyle='-', linewidth=1, alpha=0.6)
    ax.add_artist(circle)

# 绘制每个聚类的雷达图
for i in range(3):
    values = cluster_averages_normalized[i].tolist()
    values += values[:1]  # 将第一个值添加到最后一个，以闭合图形

    # 绘制线条
    ax.plot(
        angles,
        values,
        linewidth=3,
        linestyle=line_styles[i],
        color=radar_colors[i],
        label=f'类别 {i}',
        zorder=10
    )

    # 填充区域
    ax.fill(
        angles,
        values,
        color=radar_colors[i],
        alpha=alphas[i],
        zorder=5
    )

# 设置雷达图标签样式
ax.set_xticks(angles[:-1])
ax.set_xticklabels(
    labels,
    fontsize=12,
    fontweight='bold',
    color='#2c3e50'
)

# 移除刻度线和标签
ax.set_yticklabels([])
ax.set_ylim(0, 1.2)

# 添加精美的图例
legend = plt.legend(
    loc='upper right',
    bbox_to_anchor=(1.2, 1.1),
    frameon=True,
    facecolor='white',
    edgecolor='black',
    fontsize=12
)
legend.get_frame().set_linewidth(1.5)

plt.title('聚类特征雷达分析', fontsize=18, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

end_time = time.time()
execution_time = end_time - start_time
print(f"\n=== 脚本执行时间: {execution_time:.2f} 秒 ===")
```
